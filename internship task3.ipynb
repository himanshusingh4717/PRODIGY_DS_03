```python
import pandas as pd
```

```python
data = pd.read_csv("C:/Users/user/Downloads/bank.csv.zip")
data.head()
```

```python
data.drop(['default','contact','pdays','poutcome'],axis='columns',inplace=True)
```

```python
inputs = data.drop('loan',axis='columns')
target = data['loan'].map({'yes':1,'no':0})
```

```python
inputs
```

```python
target
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
age_ = LabelEncoder()
job_ = LabelEncoder()
marital_= LabelEncoder()
education_= LabelEncoder()	
balance_= LabelEncoder()	
housing_= LabelEncoder()	
month_= LabelEncoder()		
deposit_= LabelEncoder()
```

```python
inputs['age_'] = age_.fit_transform(inputs['age'])
inputs['job_'] = job_.fit_transform(inputs['job'])
inputs['marital_'] = marital_.fit_transform(inputs['marital'])
inputs['education_'] = education_.fit_transform(inputs['education'])
inputs['balance_'] = balance_.fit_transform(inputs['balance'])
inputs['housing_'] = housing_.fit_transform(inputs['housing'])
inputs['month_'] = month_.fit_transform(inputs['month'])
inputs['deposit_'] = deposit_.fit_transform(inputs['deposit'])
```

```python
input_n = inputs.drop(['age','job','marital','education','balance','housing','month','deposit'],axis='columns')
```

```python
x_train,x_test,y_train,y_test = train_test_split(input_n,target,test_size=0.2)
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```
